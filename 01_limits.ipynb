### Limits in Python

```python
from IPython.display import display, Math

display(Math(r'\lim_{{x \to \infty}} \frac{1}{x} = 0'))
```

```python
x = np.linspace(1, 100, 1000)
y = 1 / x
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y);
```

### Limit with Plotting in Python

```python
display(Math(r'\lim_{{x \to \infty}} \frac{\ln(x)}{x} = 0'))
```

```python
x = np.linspace(1, 1000, 10000)
y = np.log(x) / x
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y);
```

```python
display(Math(r'\lim_{{x \to \infty}} \frac{x^2-3x-5}{2x^2+x+1} = \frac{1}{2}'))
```

```python
x = np.linspace(1, 300, 10000)
y = (x ** 2 - 3 * x + 5) / (2 * x ** 2 + x + 1)
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y);
```

```python
display(Math(r'\lim_{{x \to 0}} \frac{\sin(x)}{x} = 1'))
```

```python
x = np.linspace(-3 * np.pi, 3 * np.pi, 1000)
y = np.sin(x) / x
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y);
```

```python
display(Math(r'\lim_{{x \to 0}} \frac{\cos(x) -1 }{x} = 0'))
```

```python
x = np.linspace(-3 * np.pi, 3 * np.pi, 1000)
y = (np.cos(x) - 1) / x
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y);
```
